## **Exercício 1: Aprovação ou reprovação**
Você estuda na USP e quer saber se foi aprovado ou não com as notas que você tem, mas seu professor utiliza o seguinte critério para cálcular a média:

1. Os alunos precisam fazer apenas uma entre P1 e P2, caso façam as duas somente a de maior nota será considerada.

2. P3 é obrigatória.

3. Calculo da média é feito por média ponderada.

<div align="center">

| Prova    | P1       | P2       | P3 |
| -------- | -------- | -------- | --- |
| **Peso** | 4        | 4        | 6   |

</div>

<br></br>

  - Se nota maior ou igual a 5, printar "Aprovado".

  - Se nota maior ou igual a 3 e menor que 5, printar "Reaval".

  - Se nota menor que 3, printar "Reprovado".

```python
# Pede ao usuário que digite as notas da P1,P2 e P3 (Prova 1, Prova 2 e Prova 3).
P1 = float(input("Digite a nota da P1: "))
P2 = float(input("Digite a nota da P2: "))
P3 = float(input("Digite a nota da P3 (obrigatória): "))

# A ideia é escolher a maior nota entre P1 e P2, já que apenas uma delas será considerada na média.
nota_maior_P1_P2 = max(P1, P2)

# Como P3 é obrigatória, ela entra sempre no cálculo
# Agora, vamos calcular a média ponderada:
# A fórmula da média ponderada para o caso em questão é:
# média = (nota_maior_P1_P2 * 4 + P3 * 6) / (4 + 6).
media = (nota_maior_P1_P2 * 4 + P3 * 6) / 10

# Print para separar a aquisição de dados da exibição do resultado.
print()

# Agora, aplicamos as regras para saber se foi aprovado, em reavaliação ou reprovado.
if media >= 5:
    print(f"Aprovado com média {media:.2f}")
elif 3 <= media < 5:
    print(f"Reaval com média {media:.2f}")
else:
    print(f"Reprovado com média {media:.2f}")
```

## **Exercício 2: Mitose**

Você é um biologo e gostaria de visualizar o quanto uma população de bactéria consegue crescer em poucos segundos.

A população de bactérias começa em uma única e **dobra de tamanho a cada segundo**.

Utilizando seus conhecimentos de programação, **crie um programa que peça o número de segundos e imprima uma pirâmide, onde cada camada tenha o número de bactérias**, após os segundo inseridos.

- #### **Exemplo:**

    input:
      4

    output:

      1

      11

      1111

      11111111

```python
# Entrada: número de segundos
segundos = int(input("Digite o número de segundos: "))

# Inicializando a população de bactérias com 1
populacao_bacterias = 1

# Inicializa o tempo inicial
tempo = 0

# Print para melhor visualização dos dados
print()

# Loop while para construir a pirâmide
while tempo < segundos:
    # Imprime a população de bactérias correspondente ao segundo atual
    print("1" * populacao_bacterias)

    # Dobra a população de bactérias a cada segundo
    populacao_bacterias *= 2

    # Incrementa o contador de tempo
    tempo += 1
```

## **Exercício 3: Mínimo Múltiplo Comum**

Seu primo pequeno vai na sua casa e vê você programando, ele, então, diz que programação não tem utilidade alguma. Você sabendo que ele está estudando sobre MMC na escola cria um **programa com Loop que permite ver qual é o MMC entre dois valores**.

Para impressiona-lo você deve encontrar o MMC entre a velocidade da Luz e do Som, que são, respectivamente,

<br></br>
<div align="center">
  $V_{luz} = 1.079.252.847$  $km/h$
<br></br>
  $V_{som} = 1233$ $km/h$
</div>

```python
# Valores dados no problema.
velocidade_luz = 1079252847  # km/h
velocidade_som = 1233        # km/h

# O MMC deve ser maior ou igual ao maior dos dois números, então começamos por ele.
mmc = velocidade_luz if velocidade_luz > velocidade_som else velocidade_som

# Usando o loop while para encontrar o MMC.
while True:
    # Verifica se 'mmc' é divisível por ambos os números.
    if mmc % velocidade_luz == 0 and mmc % velocidade_som == 0:
        # Se sim, encontramos o MMC e saímos do loop.
        break
    # Caso contrário, aumentamos o valor do mmc em 1 (ou mais) e tentamos novamente.
    mmc += 1

# Exibe o resultado do MMC.
print("O MMC entre a velocidade da luz e do som é:", mmc)
```

Podemos usar a fórmula a seguir para obter o MMC entre dois números:
<br></br>
<div align="center">
  $MMC(a, b) = \frac{a \times b}{MDC(a, b)}$
</div>

```python
# O mmc entre dois números a e b pode ser calculado seguindo a fórmula:
#                 MMC(a, b) = (a * b) / MDC(a, b)

# Valores dados no problema.
velocidade_luz = 1079252847  # km/h
velocidade_som = 1233        # km/h

# Encontrar o maior número para começar o MMC.
mmc = velocidade_luz if velocidade_luz > velocidade_som else velocidade_som

# Calcular o MDC usando um loop while. (Algoritmo de euclides)
a, b = velocidade_luz, velocidade_som
while b != 0:
    a = b
    b =  a % b
mdc = a  # Aqui temos o MDC

# Calcular o MMC usando a fórmula.
mmc = (velocidade_luz * velocidade_som) // mdc

# Exibe o resultado do MMC.
print("O MMC entre a velocidade da luz e do som é:", mmc)
```


## **Desafio 1/2: Cercado**

Você tem uma vila no Clash of Clans e gostaria de saber quantas peças serão necessárias para criar um cercado ao redor dela, a proporção da cerca será de 1:3, ou seja, **para cada 1 de largura haverá 3 de comprimento**.

Como você não sabe ainda a largura que quer e o valor que será despendido para a construção, você utilizará a programação para construir um **esboço do cercado**.

Considere que a representação do cercado será com "#".

<br></br>

  - **Exemplo:**

  input:
          
        5

  output:

        ###############
        #             #
        #             #
        #             #
        ###############

```python
# Solicita a largura do retângulo ao usuário.
largura = int(input('Digite a largura do retângulo: '))

# Define o comprimento com base na largura, sendo 3 vezes a largura.
comprimento = largura * 3

# Adiciona uma linha vazia para melhorar a visualização.
print()

# Inicializa uma variável de controle para o laço 'while'.
a = 1

# Inicia um loop que percorre cada linha do retângulo.
while a <= largura:

    # Se for a primeira linha ou a última linha, imprime uma linha completa com '#'.
    if a == 1 or a == largura:
        print(f'{"#"*comprimento}')
    else:
        # Caso contrário, imprime uma linha com bordas e espaços no meio.
        vazio = comprimento - 2  # Define o número de espaços entre as bordas.
        print(f'#{" "*vazio}#')

    # Incrementa a variável de controle para avançar para a próxima linha.
    a += 1
```

## **Desafio 2/2: Recibo de compra**
Crie um programa que peça produto, número de série e valor, respeitando as instruções abaixo. Caso não siga o padrão, deve-se printar uma mensagem de erro.
1. **Produto:** Somente letras; Sem espaços vazios; Sem limite para tamanho.
2. **Número de série:** Não podem haver simbolos ou letras; Sem espaços vazios; Devem ter exatamente 4 números inteiros.
3. **Valor:** Podem ser números inteiros ou decimais; Todo Valor terá duas casas decimais após a virgula.

- #### Exemplo de recibo:

  Produto: Shampoo

  Nº de série: 1234

  Valor: 26.34

- #### Exemplo de mensagem de erro:

  input:
    
      "Camise6a"

  output:
  
      "ERRO: produto incorreto, escreva o nome do produto somente com letras e sem espaços vazios."

```python
# Pede ao usuário que insira o nome do produto.
produto = input("Digite o nome do produto (somente letras, sem espaços): ")

# Verifica se o produto contém apenas letras e não possui espaços.
while not produto.isalpha():  # isalpha retorna true se todos os caracteres da string forem letras do alfabeto.
    # Perceba que o fluxo de execução só chegará aqui se o nome do produto inserido pelo usuário for inválido de acordo com as regras que definimos.
    # Caso o fluxo de execução chegue até aqui, exibe a mensagem de erro e pede novamente que o usuário insira o nome do produto.
    print("ERRO: produto incorreto, escreva o nome do produto somente com letras e sem espaços vazios.")
    produto = input("Digite o nome do produto (somente letras, sem espaços): ")

# Pede ao usuário que insira o número de série do produto.
numero_serie = input("Digite o número de série (4 números, sem letras ou símbolos): ")

# Verifica se o número de série tem exatamente 4 dígitos e se esses dígitos são apenas números.
while not (len(numero_serie) == 4 and numero_serie.isdigit()):  # Verifica tamanho e se são números
    # Caso o fluxo de execução chegue até aqui, exibe a mensagem de erro e pede novamente que o usuário insira o número de série do produto.
    print("ERRO: número de série incorreto, ele deve conter exatamente 4 números inteiros.")
    numero_serie = input("Digite o número de série (4 números, sem letras ou símbolos): ")

# Pede ao usuário que insira o valor do produto.
valor = input("Digite o valor (com duas casas decimais, separado por ponto '.'): ")

# Verifica se o valor é um número válido com exatamente duas casas decimais.
while True:
    partes = valor.split(".")  # Divide a string por ".".

    # Checa se temos duas partes, se a parte inteira e decimal são números e se a decimal tem 2 dígitos.
    if len(partes) == 2 and partes[0].isdigit() and partes[1].isdigit() and len(partes[1]) == 2:
        break
    else:
        print("ERRO: valor incorreto, ele deve ser um número com duas casas decimais.")
        valor = input("Digite o valor (com duas casas decimais, separado por ponto '.'): ")

# Se tudo estiver correto, imprime o recibo.
print(f"\nProduto: {produto}")
print(f"Nº de série: {numero_serie}")
print(f"Valor: {valor}")
```
