## Exerc√≠cio 1: Planejamento de Viagem (Convers√£o de Moedas)


Voc√™ est√° planejando uma viagem internacional e precisa converter seu dinheiro para outra moeda.

1. **Crie um programa que converta reais (R\$) para d√≥lares (USD)**. O programa deve pedir ao usu√°rio a quantidade em reais e usar uma taxa de c√¢mbio fornecida para calcular o valor correspondente em d√≥lares.
   - **Exemplo**: Se a taxa de c√¢mbio for 1 USD = 5,30 BRL e voc√™ tiver R$ 1.000, qual ser√° o valor em d√≥lares?

2. **Modifique o programa** para que o usu√°rio converta reais para euros tamb√©m (use 1 EUR = 5,80 BRL).


```python
# Pergunta ao usu√°rio qual √© a taxa de c√¢mbio entre real e dolar.
taxa_de_cambio_dolar = input("Qual √© a taxa de c√¢mbio que ser√° utilizada entre real e dolar? ")

# Pergunta ao usu√°rio qual √© a quantidade de reais que ser√£o convertidos para d√≥lares.
quantidade_em_reais = input("Qual √© a quantidade de reais que ser√° convertida em d√≥lares? ")

# Converte o valor em reais para um valor em dolar.
# Observe que, antes de realizar os c√°lculos, os valores "quantidade_em_reais" e "taxa_de_cambio" foram convertidos para float.
valor_em_dolar = float(quantidade_em_reais) / float(taxa_de_cambio_dolar)

# Pula uma linha para separar a aquisi√ß√£o dos dados da exibi√ß√£o dos resultados.
print("\n")

# Imprime o valor convertido em dolar.
print(f"O valor em d√≥lares √©: {valor_em_dolar:.2f} $")

# ----------------------------------------------------------------------------------------------------------------------------

# Seta o valor da taxa de c√¢mbio entre real e euro.
taxa_de_cambio_euro = 5.8

# Converte o valor em reais para um valor em euro.
valor_em_euro = float(quantidade_em_reais) / float(taxa_de_cambio_euro)

# Imprime o valor convertido em euro. Observe que o n√∫mero de casas decimais da sa√≠da √© formatado usando ":.2f".
print(f"O valor em euros √©: {valor_em_euro:.2f} ‚Ç¨")
```

## Exerc√≠cio 2: C√°lculo do Consumo de Combust√≠vel


Voc√™ est√° indo em uma viagem de carro e deseja saber quanto combust√≠vel gastar√°.

1. **Escreva um programa que calcule o consumo de combust√≠vel baseado na dist√¢ncia a ser percorrida e no consumo m√©dio do carro (km por litro)**. O programa deve perguntar ao usu√°rio a dist√¢ncia em quil√¥metros e o consumo m√©dio do carro.
   - **Exemplo**: Se a dist√¢ncia √© de 300 km e o carro faz 12 km/L, quantos litros de combust√≠vel ser√£o necess√°rios?

2. **Adicione um c√°lculo de custo**, onde o programa pergunta tamb√©m o pre√ßo do litro de combust√≠vel e exibe o custo total da viagem.


```python
# Obs: Veja que os 3 inputs abaixo s√£o diretamente convertidos para float antes de serem salvos em uma vari√°vel.

# Pergunta ao usu√°rio qual ser√° a dist√¢ncia a ser percorrida com o carro.
distancia_a_ser_percorrida = float(input("Qual √© a dist√¢ncia (em quil√¥metros) a ser percorrida ? "))

# Pergunta ao usu√°rio qual √© o consumo m√©dio do carro.
consumo_medio_do_carro = float(input("Qual √© o consumo m√©dio do carro (em quil√¥metros por litro) ? "))

# Pergunta ao usu√°rio qual √© o pre√ßo do litro de combust√≠vel.
custo_por_litro_de_combustivel = float(input("Qual √© o custo do litro de combust√≠vel (em R$) ? "))

# Calcula quantos litros de combust√≠vel ser√£o gastos na viagem.
litros_de_combustivel_necessarios = distancia_a_ser_percorrida / consumo_medio_do_carro

# Calcula o custo total da viagem.
custo_total_da_viagem = litros_de_combustivel_necessarios*custo_por_litro_de_combustivel

# Pula uma linha para separar a aquisi√ß√£o dos dados da exibi√ß√£o dos resultados.
print("\n")

# Exibe a quantidade de litros de combust√≠vel que ser√£o necess√°rios na viagem.
print(f"Ser√£o necess√°rios {litros_de_combustivel_necessarios:.2f} litros de combust√≠vel para completar a viagem")

# Exibe o custo total da viagem.
print(f"O custo total da viagem √©: {custo_total_da_viagem:.2f} R$")
```

## Exerc√≠cio 3: Gerador de Senhas Personalizadas (Manipula√ß√£o de Strings)


Voc√™ est√° criando um sistema que gera senhas personalizadas para os usu√°rios com base em suas informa√ß√µes pessoais. As senhas precisam ser √∫nicas e seguras, mas tamb√©m f√°ceis de lembrar.

1. **Crie um programa que pe√ßa ao usu√°rio seu nome e uma data**. O programa deve gerar uma senha usando a seguinte l√≥gica:
   - As 3 primeiras letras do nome.
   - Os dois primeiros d√≠gitos do ano.
   - O n√∫mero de caracteres do nome completo (incluindo espa√ßos).

   **Exemplo**: Para um usu√°rio chamado "Ana Beatriz Costa" e "15/08/1990", a senha seria: `Ana19 18`.

2. **Modifique o programa para transformar as letras da senha gerada em letras mai√∫sculas e adicionar um caractere especial "!" no final da senha**.

   **Exemplo modificado**: A senha final seria: `ANA19 18!`.


```python
# Pede ao usu√°rio que insira o seu nome.
nome = input("Qual √© o seu nome? ")

# Pede ao usu√°rio que insira uma data no formato "DD/MM/AAAA".
data = input("Insira uma data no formato 'DD/MM/AAAA': ")

# Precisamos das 3 primeiras letras do nome inserido pelo usu√°rio, podemos fazer isso obtendo uma letra por vez, veja:
iniciais_nome = nome[0] + nome[1] + nome[2]  # Obtemos e concatenamos as 3 primeiras letras do nome do usu√°rio.
# Obs: Essa opera√ß√£o poderia ser simplificada para iniciais_nome = nome[:3]       (mais detalhes nas pr√≥ximas aulas =])

# Precisamos dos dois primeiros d√≠gitos do ano, como o ano est√° no formato 'DD/MM/AAAA' basta que peguemos os d√≠gitos da posi√ß√£o 6 e 7.
# Obs: Em computa√ß√£o, come√ßamos a contar do 0 e n√£o do 1. Ou seja, temos a seguinte rela√ß√£o entre posi√ß√£o e d√≠gito para a data:
#       D - D√≠gito 0
#       D - D√≠gito 1
#       / - D√≠gito 2
#       M - D√≠gito 3
#       M - D√≠gito 4
#       / - D√≠gito 5
#       A - D√≠gito 6
#       A - D√≠gito 7
#       A - D√≠gito 8
#       A - D√≠gito 9
digitos_ano = data[6] + data[7] # Obtemos e concatenamos os 2 primeiros d√≠gitos do ano.

# Cria a senha no formato especificado (observe que um espa√ßo deve separar a jun√ß√£o das vari√°veis "iniciais_nome" e "digitos_ano", do n√∫mero
# de caracteres da vari√°vel "nome").
senha = iniciais_nome + digitos_ano + " " + str(len(nome))

# Pula uma linha para separar a aquisi√ß√£o dos dados da exibi√ß√£o dos resultados.
print("\n")

# Exibe a senha sem modifica√ß√µes.
print("Senha sem modifica√ß√µes: " + senha)

# Transforma as letras da senha gerada em letras mai√∫sculas e adiciona um caractere especial "!" no final da senha.
senha_modificada = senha.upper() + "!"

# Exibe a senha modificada.
print("Senha modificada: " + senha_modificada)
```

## Exerc√≠cio 4: C√°lculo de Juros Simples (Aplica√ß√£o Financeira)


Voc√™ est√° pensando em investir uma quantia em um banco e quer saber quanto receber√° ao final de um per√≠odo.

1. **Escreva um programa que calcule o valor final de uma aplica√ß√£o de juros simples**, com base no capital inicial, taxa de juros e tempo de investimento. A f√≥rmula √© $M = C 	\cdot (1 + i)^{t}$, onde:
   - \( M \) √© o montante final,
   - \( C \) √© o capital inicial,
   - \( i \) √© a taxa de juros anual,
   - \( t \) √© o tempo (em anos).
   - **Exemplo**: Um investimento de R$ 1.000 a uma taxa de 5% ao ano por 3 anos.

2. **Permita que o usu√°rio escolha o per√≠odo de tempo (em meses ou anos)** e fa√ßa a convers√£o correta no c√°lculo.


```python
# Pede ao usu√°rio que insira o valor do capital inicial.
capital_inicial = float(input("Qual √© o valor do capital inicial? "))

# Pergunta ao usu√°rio se ele quer que a taxa de juros seja mensal ou anual.
periodo_de_tempo = input("Digite 'A' para taxa anual ou 'M' para taxa mensal: ")

# Pede ao usu√°rio que insira o valor da taxa de juros no formato decimal (isto √© 5% seria 0.05).
taxa_de_juros = float(input("Qual √© o valor da taxa de juros? (insira tal valor no formato decimal, isto √©, 5% seria 0.05): "))

# Pede ao usu√°rio que insira o tempo de investimento.
tempo_de_investimento = float(input("Qual √© o tempo de investimento? "))

# Usa uma condicional para ajustar o valor da taxa de juros.
if(periodo_de_tempo == "A"):
  taxa_de_juros = taxa_de_juros
else:
  taxa_de_juros = (1 + taxa_de_juros)**(1/12)-1 # F√≥rmula usada: taxa_mensal = (1 + taxa_anual)^(1/12) - 1.

# Calcula o montante final do investimento.
montante_final = capital_inicial * (1 + taxa_de_juros)**(tempo_de_investimento)

# Pula uma linha para separar a aquisi√ß√£o dos dados da exibi√ß√£o do resultado.
print("\n")

# Exibe o valor do montante final.
print(f"O valor do montante final ser√°: {montante_final:.2f}R$")
```

# Desafio: Verifica√ß√£o de ID V√°lido


Voc√™ est√° trabalhando com um sistema de verifica√ß√£o de identidade, e precisa validar se um n√∫mero de identifica√ß√£o segue uma regra espec√≠fica.

1. **Crie um programa que valide se um n√∫mero de CPF √© v√°lido**, baseado no formato "XXX.XXX.XXX-XX". O programa deve verificar se a entrada tem o n√∫mero correto de d√≠gitos e o formato correto (n√£o √© necess√°rio validar os d√≠gitos verificadores).
   - **Exemplo**: O programa deve aceitar "123.456.789-00" e rejeitar "12345678900".

2. **Adicione uma funcionalidade** onde o programa tamb√©m valida um CNPJ no formato "XX.XXX.XXX/0001-XX".


```python
# Vamos come√ßar validando um CPF
# Sabemos que o CPF √© v√°lido se estiver no formato XXX.XXX.XXX-XX. Portanto, devemos realizar as seguintes verifica√ß√µes:
#       1) O CPF recebido como entrada possui 14 digitos ?
#       2) Os d√≠gitos das posi√ß√µes 3 e 7 s√£o "." ?
#       3) O d√≠gito da posi√ß√£o 11 √© um "-" ?

# Pedimos ao usu√°rio para que digite o CPF que ele deseja verificar a validade
cpf = input("Digite o CPF que voc√™ deseja verificar a validade: ")

# Criamos uma vari√°vel booleana para guardar se o CPF inserido pelo usu√°rio tem o n√∫mero correto de d√≠gitos ou n√£o (isto √©, 14 d√≠gitos).
condicao_1 = len(cpf) == 14

# Criamos uma vari√°vel booleana para guardar se os d√≠gitos das posi√ß√µes 3 e 7 s√£o "."
condicao_2 = cpf[3] == "." and cpf[7] == "."

# Criamos uma vari√°vel booleana para guardar se o d√≠gito da posi√ß√£o 11 √© um "-"
condicao_3 = cpf[11] == "-"

# Pula uma linha para separar a aquisi√ß√£o dos dados da exibi√ß√£o do resultado.
print("\n")

# Criamos uma condicional para ajudar a mostrar se o CPF inserido √© v√°lido ou n√£o.
if(condicao_1 and condicao_2 and condicao_3):
  print("CPF v√°lido")
else:
  print("CPF inv√°lido")
```

```python
# Vamos validar agora um CNPF
# Sabemos que o CNPF √© v√°lido se estiver no formato XX.XXX.XXX/0001-XX. Portanto, devemos realizar as seguintes verifica√ß√µes:
#       1) O CNPJ recebido como entrada possui 18 digitos ?
#       2) Os d√≠gitos das posi√ß√µes 2 e 6 s√£o "." ?
#       3) O d√≠gito da posi√ß√£o 10 √© um caractere "/" ?
#       4) A sequ√™ncia dos caracteres nas posi√ß√µes 11, 12, 13, 14 √© igual √† 0001 ?
#       5) O d√≠gito da posi√ß√£o 15 √© um caractere "-" ?

# Pedimos ao usu√°rio para que digite o CNPJ que ele deseja verificar a validade
cnpj = input("Digite o CNPJ que voc√™ deseja verificar a validade: ")

# Criamos uma vari√°vel booleana para guardar se o CNPJ inserido pelo usu√°rio tem o n√∫mero correto de d√≠gitos ou n√£o (isto √©, 18 d√≠gitos).
condicao_1 = len(cnpj) == 18

# Criamos uma vari√°vel booleana para guardar se os d√≠gitos das posi√ß√µes 3 e 7 s√£o "."
condicao_2 = cnpj[2] == "." and cnpj[6] == "."

# Criamos uma vari√°vel booleana para guardar se o d√≠gito da posi√ß√£o 11 √© um "-"
condicao_3 = cnpj[10] == "/"

# Criamos uma vari√°vel booleana para guardar se a sequ√™ncia dos caracteres nas posi√ß√µes 12, 13, 14 e 15 √© igual √† 0001
condicao_4 = cnpj[11] + cnpj[12] + cnpj[13] + cnpj[14] == "0001"

# Criamos uma vari√°vel booleana para guardar se o d√≠gito da posi√ß√£o 16 √© um "-"
condicao_5 = cnpj[15] == "-"

# Pula uma linha para separar a aquisi√ß√£o dos dados da exibi√ß√£o do resultado.
print("\n")

# Criamos uma condicional para ajudar a mostrar se o CNPJ inserido √© v√°lido ou n√£o.
if(condicao_1 and condicao_2 and condicao_3 and condicao_4 and condicao_5):
  print("CNPJ v√°lido")
else:
  print("CNPJ inv√°lido")
```
